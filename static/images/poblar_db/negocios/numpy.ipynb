```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1️⃣ Datos crudos del sensor
# -------------------------------
# Estoy simulando datos de temperatura que podrían venir de un sensor real.
# Incluyo valores nulos, fuera de rango y algo de ruido para hacer el proceso realista.
datos_crudos = [21.3, 22.1, None, 25.0, 300.0, 23.5, np.nan, 19.8, 21.7, -5.0, 22.9]
lecturas = np.array(datos_crudos, dtype=np.float64)

# Gráfico inicial de los datos crudos
plt.figure(figsize=(10, 5))
plt.plot(lecturas, marker='o', linestyle='--', label="Datos crudos")
plt.title("Paso 1: Datos crudos del sensor")
plt.xlabel("Muestra")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------
# 2️⃣ Eliminar o reemplazar valores nulos
# -------------------------------
# Primero elimino los NaN para que no afecten mis cálculos iniciales
lecturas_sin_nan = lecturas[~np.isnan(lecturas)]

# Gráfico después de quitar NaN
plt.figure(figsize=(10, 5))
plt.plot(lecturas_sin_nan, marker='o', linestyle='--', color='orange', label="Sin NaN")
plt.title("Paso 2: Eliminación de valores nulos")
plt.xlabel("Muestra")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------
# 3️⃣ Filtrar valores fuera de rango
# -------------------------------
# Sé que la temperatura ambiente debe estar entre 0 y 50°C
lecturas_filtradas = lecturas_sin_nan[(lecturas_sin_nan >= 0) & (lecturas_sin_nan <= 50)]

# Gráfico después del filtrado
plt.figure(figsize=(10, 5))
plt.plot(lecturas_filtradas, marker='o', linestyle='--', color='green', label="Filtrado 0°C - 50°C")
plt.title("Paso 3: Filtrado de valores fuera de rango")
plt.xlabel("Muestra")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------
# 4️⃣ Conversión de unidades (opcional)
# -------------------------------
# Supongamos que este paso implica convertir °C a °F para un cliente extranjero
lecturas_fahrenheit = (lecturas_filtradas * 9/5) + 32

# Gráfico con datos convertidos
plt.figure(figsize=(10, 5))
plt.plot(lecturas_fahrenheit, marker='o', linestyle='--', color='red', label="Convertido a °F")
plt.title("Paso 4: Conversión de unidades (°C → °F)")
plt.xlabel("Muestra")
plt.ylabel("Temperatura (°F)")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------
# 5️⃣ Suavizado de datos (Promedio móvil)
# -------------------------------
# Creo una función para suavizar los datos con promedio móvil
def promedio_movil(datos, ventana=3):
    return np.convolve(datos, np.ones(ventana)/ventana, mode='valid')

lecturas_suavizadas = promedio_movil(lecturas_filtradas, ventana=3)

# Gráfico del suavizado
plt.figure(figsize=(10, 5))
plt.plot(lecturas_filtradas, marker='o', linestyle='--', color='green', label="Original filtrado")
plt.plot(range(1, len(lecturas_suavizadas)+1), lecturas_suavizadas, marker='o', color='purple', label="Suavizado")
plt.title("Paso 5: Suavizado de datos con promedio móvil")
plt.xlabel("Muestra")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid(True)
plt.show()
```
